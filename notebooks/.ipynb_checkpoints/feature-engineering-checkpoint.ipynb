```python
import pandas as pd
import numpy as np
```

```python
app_usage = pd.read_csv("app_usage.csv")
beacon = pd.read_csv("beacon_data.csv")
product = pd.read_csv("product_order.csv")
```

```python
# encode session id as string
app_usage['session_id'] = app_usage['session_id'].astype(str)
product['session_id'] = app_usage['session_id'].astype(str)
```

```python
app_usage = app_usage.sort_values(by=['user_customer_id'], na_position='last').drop_duplicates(subset=['session_id'])
```

```python
app_usage.shape
```

```python
product.head()
```

```python
product_features_cols = ['session_id', 'price', 'quantity', 'total_cost']
```

```python
product_features = product[product_features_cols].groupby(by=['session_id']).sum()
```

```python
product_features
```

##### Get app_usage and product data together

```python
data = app_usage.merge(product_features, on=['session_id'], how='left')
data
```

```python
# beacon data at user level - this should be unique for beacon.
print(beacon.shape)
print(beacon.user_id.nunique())
```

```python
beacon_features_cols = set(beacon.columns) - set(app_usage.columns)
beacon_features_cols = beacon_features_cols.union({'user_id', 'event_date'})
beacon_features_cols = list(beacon_features_cols)
```

```python
beacon_features = beacon[beacon_features_cols].copy(deep=True)
beacon_features
```

###### We want unique users. But there are multiple users and the features here are binary so cannot just be aggregated with a sum or a count function. Hence, time to investigate a bit more on whether or not a single user has 1 and 0 for the values accross these features. 

```python
beacon_features_users = beacon_features.groupby(by=['user_id']).count().sort_values(by=['entered_store'], ascending=False)
beacon_features_users
```

Lets look at the first three users

```python
beacon[beacon.user_id == 'effbc254ba576278fa5d064491958314']
```

Clearly the same user has multiple 'sessions' but these aren't provided ID's in the beacon data. For now, it might just do to take the user activity based on the latest event_date and drop the rest. Then join that to the main table. 

```python
beacon_features = beacon_features.sort_values(
    by=['user_id', 'event_date'], ascending=False, na_position='last').drop_duplicates(subset=['user_id'])
```

```python
beacon_features
```

#### Put all of the tables together

```python
data = data.merge(beacon_features, on=['user_id', 'event_date'], how='left')
data
```

```python
# Getting rid of ids and columns that aren't going to be used for model effectively (e.g. transaction_value as this
# will lead to data leakage in the case where we are predicting completed_transaction)
X_cols = ['event_date', 'session_timestamp', 'event_wday', 'customer_type',
       'session_duration_microseconds', 'signed_up', 'pages_visited', 'app_removed',
       'personalise_store_pageviews', 'purposeful_shelves_pageviews',
       'digi_me_pageviews', 'homepage_duration_microseconds', 'add_to_cart',
       'scan_qr', 'notification_open', 'notification_receive', 'price',
       'quantity', 'total_cost', 'entered_store', 'visited_purposeful_shelves',
       'entered_dm', 'visited_digi_me', 'visited_personalised_store',
       'entered_welcome', 'entered_purposeful_shelves']

y_col = ['session_id', 'completed_transaction']
```

Encoding week days numerically

```python
print(data.event_wday.unique())
event_wday_recode = {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}
data['event_wday'] = data['event_wday'].replace(event_wday_recode)
```

Encoding other categorical variables numerically

```python
print(data.customer_type.unique())
customer_type_recode = {'Returning Visitor': 0, 'New Visitor': 1}
data.customer_type = data.customer_type.replace(customer_type_recode)
```

Normalization of some features

```python
# price, quantity and total_cost need to be normalised
from sklearn import preprocessing

data = data.replace({np.nan: 0})

minmax = preprocessing.MinMaxScaler(feature_range=(0,1))
data[['price', 'quantity', 
      'total_cost', 'session_duration_microseconds',
     'homepage_duration_microseconds']] = minmax.fit_transform(data[['price', 'quantity', 
                                                                     'total_cost', 'session_duration_microseconds',
                                                                    'homepage_duration_microseconds']])
```

```python
print(data['entered_store'].unique()) 
print(data['visited_purposeful_shelves'].unique())
print(data['entered_dm'].unique()) 
print(data['visited_digi_me'].unique()) 
print(data['visited_personalised_store'].unique())
print(data['entered_welcome'].unique()) 
print(data['entered_purposeful_shelves'].unique())
```

```python
data['event_date']
```

Encoding month as well as weekdays using a cyclical function

```python
# Date and time feature engineering - only keep month from event_date and transform month and time to cyclical function.
import math
# normalize month to match the 0- 1/2pi cycle
data['month'] = pd.to_datetime(data['event_date'], dayfirst=True).dt.month
data['month_norm'] = math.pi/2 * data['month']/data['month'].max()
data['cos_month'] = np.cos(data['month_norm'])
```

```python
data[['month', 'month_norm', 'cos_month']]
```

```python
# sample cyclical encoding of weekdays. 

data['event_wday_norm'] = 1/2 * math.pi * data['event_wday']/data['event_wday'].max()
data['cos_wday'] = np.cos(data['event_wday_norm'])
```

```python
data[['event_wday', 'event_wday_norm', 'cos_wday']]
```
