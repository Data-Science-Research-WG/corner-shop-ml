###### Investigating app usage join with product

```python
product['session_timestamp'] = product['time']
```

```python
app_product = pd.merge(app_usage, product, on=['session_timestamp', 'session_id'], how='left')
```

```python
app_product
```

Only investigate session_ids that have an order against them

```python
app_product_order = app_product.dropna(subset=['order_id'])
app_product_order
```

```python
app_product_order = app_product_order.copy(deep=True)
app_product_order['temp_id'] = [i for i in range(0, app_product_order.shape[0])]
```

```python
session_id_count = app_product_order[['temp_id', 'session_id']].groupby(by=['session_id']).count().sort_values('temp_id', ascending=False)
session_id_count
```

Take the most populated session ids to see how the join affects them

```python
session_id_1 = session_id_count.index[0]
session_id_2 = session_id_count.index[1]
session_id_3 = session_id_count.index[2]
session_id_4 = session_id_count.index[3]
session_id_5 = session_id_count.index[4]
session_ids_check = [session_id_1, session_id_2, session_id_3, session_id_4, session_id_5]
```

### Other observations & issues to flag related to Data Quality

The Data quality or recording is not great for those for which purchases were made (e.g. not set item names).

Why have two records of the same item being ordered been assigned to separate item_ids? This is an issue because while we could get the product order table aggregated to a session level by summing up the price, quantity and cost, with separate item_ids for the same item being bought (and assigned quantity = 1, instead of 2) we would potentially be duplicating records and counts - hence unreliable data. 

###### The general relationship is that a single session can have multiple orders (order_ids) and these orders can have multiple items (item_id). 

```python
sessions_dfs_list[2]
```

```python
sessions_dfs_list[2].drop_duplicates(subset=['order_id','session_timestamp', 'item_id'])
```

```python
sessions_dfs_list[3]
```

```python
sessions_dfs_list[3].drop_duplicates(subset=['order_id', 'item_id', 'session_timestamp'])
```

```python
sessions_dfs_list[4]
```

#### A possible avenue with Product Order table

1. Aggregate the Data to a session level by summing across price, quantity and total_cost. This ignores the multiple order ids and items that are linked to the session. 
    - An advantage of this is that we don't have to worry about (not set) types of items either. 
    - Although, doing this could enable DQ issues to seep in. From the examples it seems like the same items are being assigned different item_ids event though they could potentially have been summed up to a single record by quantity, price and total cost.
